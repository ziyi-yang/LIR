```python
from collections import defaultdict
import numpy as np
from tqdm import tqdm
from sklearn.metrics import average_precision_score
import copy
import json
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

import tensorflow_datasets as tfds
import tensorflow_text as text  # Import TF.text ops
import tensorflow.compat.v2 as tf
import tensorflow_hub as hub

# Make sure that TF Datasets library has wiki40b
assert tfds.__version__ == "4.0.1"
```

## Load Multilingual Sentences from Wiki40B

```python
# Use 10,000 sentences to extract the language information
sents_limit = 10000

lan_list = ["ar", "de", "el", "en", "es", "hi", "ru", "th", "tr", "vi", "zh-cn"]
text_wiki = defaultdict(list)
for lan in lan_list:
    dataset = tfds.load(f"wiki40b/{lan}")["train"]
    dataset = tfds.as_numpy(dataset)
    cnt = 0
    for example in dataset:
        for text_sub in example["text"].decode("utf-8").split("_START_PARAGRAPH_"):
            if "_START_" not in text_sub:
                text_list = text_sub.split("_NEWLINE_")
                text_list = [a_.replace("\n", "") for a_ in text_list if len(a_) >= 10]
                if text_list:
                    text_wiki[lan].extend(text_list)
                    cnt += len(text_list)
        if cnt >= sents_limit:
            break
text_wiki["zh"] = text_wiki.pop("zh-cn")
```

# Extract Language Information via SVD

```python
# Here we use model X_X as an example
model_name = "X_X"

# Load LAReQA encoder
loaded = hub.load(f"https://tfhub.dev/google/LAReQA/mBERT_{model_name}/1")

question_encoder = loaded.signatures["query_encoder"]
response_encoder = loaded.signatures["response_encoder"]
```

```python
# key: language name, val: language information of size d*20, where d is the representation size 
lan_info = {}

# batch size
b_size = 128
for k, v in tqdm(text_wiki.items()):
    v = np.asarray(v)
    emb = []
    for i in range(0, len(v), b_size):
        emb.append(question_encoder(input=tf.constant(v[i:i+b_size]))["outputs"].numpy())
    emb = np.concatenate(emb)
    u, _, _ = np.linalg.svd(np.transpose(emb), full_matrices=False)
    lan_info[k] = u[:, :20]

    lan_info["zh"] = lan_info.pop("zh-cn")
```

# Apply LIR to LAReQA XQuAD-r dataset

```python
lan_list = ["ar", "de", "el", "en", "es", "hi", "ru", "th", "tr", "vi", "zh"]

questions = {}
for lan in lan_list:
    questions[lan] = defaultdict(list)

# key: qid, embeds
candidates = defaultdict(list)

# memorize the corresponding index of each language in "candidates"
index = {}
```

```python
# load dataset
dataset_name = "xquad-r"

start_ = 0
candidates["sentences"] = []
candidates["context"] = []
candidates["qid"] = []

for lan in lan_list:
    questions[lan]["qid"] = []
    questions[lan]["question"] = []
    f = open(f"{dataset_name}/{lan}.json")
    data = json.load(f)["data"]
    for entry in data:
        for para in entry["paragraphs"]:
            n_sents = len(para["sentences"])
            context = [para["context"]]*n_sents            
            qid_list = [qs["id"] for qs in para["qas"]]
            q_list = [qs["question"] for qs in para["qas"]]
            sent_qid_list = [set() for _ in range(n_sents)]
            for qs in para["qas"]:
                a_start = qs["answers"][0]["answer_start"]
                for i in range(n_sents):
                    if a_start >= para["sentence_breaks"][i][0] and a_start <= para["sentence_breaks"][i][1]:
                        sent_qid_list[i].add(qs["id"])
                        break
            candidates["sentences"].extend(para["sentences"])
            candidates["context"].extend(context)
            candidates["qid"].extend(sent_qid_list)
            questions[lan]["qid"].extend(qid_list)
            questions[lan]["question"].extend(q_list)
    index[lan] = [start_, len(candidates["sentences"])]
    start_ = len(candidates["sentences"])
    f.close()
```

```python
# encode dataset

for lan in lan_list:
    for k, v in questions[lan].items():
        if k != "qid":
            questions[lan][k] = np.asarray(v)

for k, v in candidates.items():
    if k != "qid":
        candidates[k] = np.asarray(v)

for lan in tqdm(lan_list):
    q_lan = questions[lan]["question"]
    for i in range(0, len(q_lan), b_size):
        embeds_ = question_encoder(
            input=tf.constant(q_lan[i:i+b_size]))["outputs"]
        questions[lan]["embeds"].append(embeds_.numpy())

for i in tqdm(range(0, len(candidates["sentences"]), b_size)):
    embeds_ = response_encoder(
        input=tf.constant(candidates["sentences"][i:i+b_size]),
        context=tf.constant(candidates["context"][i:i+b_size]))["outputs"]
    candidates["embeds"].append(embeds_.numpy())

for lan in lan_list:
    questions[lan]["embeds"] = np.concatenate(questions[lan]["embeds"])
candidates["embeds"] = np.concatenate(candidates["embeds"])
```

```python
def eval_lareqa(q_dict, cand_dict, languages):
    """
    Compute the averaged mAP across all languages.
    """
    score = 0
    cnt = 0
    result = {}
    for lan in tqdm(languages):
        cnt_lan = 0
        score_lan = 0
        dots = np.matmul(q_dict[lan]["embeds"], np.transpose(cand_dict["embeds"]))
        n_q = len(q_dict[lan]["qid"])
        for i in range(n_q):
            qid = q_dict[lan]["qid"][i]
            labels = [1 if qid in id_set else 0 for id_set in cand_dict["qid"]]
            tmp = average_precision_score(labels, dots[i, :])
            score += tmp
            score_lan += tmp
            cnt += 1
            cnt_lan += 1
        result[lan] = score_lan/cnt_lan
    result["all"] = score/cnt
    return result

def lir(embed, c, r = 1):
    # Remove the language information from embeddings
    # r is how many components to remove
    proj = np.matmul(embed, c[:, :r])
    return embed - np.matmul(proj, np.transpose(c[:, :r]))
```

```python
candidates_lir = copy.deepcopy(candidates)
questions_lir = copy.deepcopy(questions)

# e.g. here we remove one component
num_remove = 1

for lan, (s_, e_) in index.items():
    candidates_lir["embeds"][s_:e_, :] = lir(candidates_lir["embeds"][s_:e_, :], lan_info[lan], r=num_remove)

for lan in tqdm(lan_list):
    questions_lir[lan]["embeds"] = lir(questions_lir[lan]["embeds"], lan_info[lan], r=num_remove)

result_no_lir = eval_lareqa(questions, candidates, lan_list)
result_lir = eval_lareqa(questions_lir, candidates_lir, lan_list)
```

```python
map_no_lir = result_no_lir["all"]
print(f"Before applying LIR, averaged mAP is: {map_no_lir}")
map_lir = result_lir["all"]
print(f"After applying LIR (removing {num_remove} components), averaged mAP is: {map_lir}")
```

```python
candidates_lir = copy.deepcopy(candidates)
questions_lir = copy.deepcopy(questions)

# e.g. here we remove one component
num_remove = 2

for lan, (s_, e_) in index.items():
    candidates_lir["embeds"][s_:e_, :] = lir(candidates_lir["embeds"][s_:e_, :], lan_info[lan], r=num_remove)

for lan in tqdm(lan_list):
    questions_lir[lan]["embeds"] = lir(questions_lir[lan]["embeds"], lan_info[lan], r=num_remove)

result_lir = eval_lareqa(questions_lir, candidates_lir, lan_list)

map_lir = result_lir["all"]
print(f"After applying LIR (removing {num_remove} components), averaged mAP is: {map_lir}")
```

We find that removing just one component from candidates_lir can further improve the performance.

```python
candidates_lir = copy.deepcopy(candidates)
questions_lir = copy.deepcopy(questions)

# e.g. here we remove one component
num_remove = 2

for lan, (s_, e_) in index.items():
    # r=1 is not a typo
    candidates_lir["embeds"][s_:e_, :] = lir(candidates_lir["embeds"][s_:e_, :], lan_info[lan], r=1)

for lan in tqdm(lan_list):
    questions_lir[lan]["embeds"] = lir(questions_lir[lan]["embeds"], lan_info[lan], r=num_remove)

result_lir = eval_lareqa(questions_lir, candidates_lir, lan_list)

map_lir = result_lir["all"]
print("Removing juist 1 component from candidates embeddings.")
print(f"After applying LIR (removing {num_remove} components), averaged mAP is: {map_lir}")
```

# LIR Visualization

```python
def pca_plot(q_en_emb, c_en_emb, q_zh_emb, c_zh_emb):
    pca = PCA(n_components=2)
    pca.fit(np.concatenate([q_en_emb, c_en_emb, q_zh_emb, c_zh_emb]))
    q_en_emb = pca.transform(q_en_emb)
    c_en_emb = pca.transform(c_en_emb)    
    q_zh_emb = pca.transform(q_zh_emb)    
    c_zh_emb = pca.transform(c_zh_emb)
    
    f = plt.figure(figsize=(4, 3))
    alpha = 0.4
    plt.scatter(c_en_emb[:, 0], c_en_emb[:, 1], c="lightgreen", alpha=alpha)
    plt.scatter(q_en_emb[:, 0], q_en_emb[:, 1], c="cornflowerblue", alpha=alpha)
        
    plt.scatter(c_zh_emb[:, 0], c_zh_emb[:, 1], c="darkorange", alpha=alpha)
    plt.scatter(q_zh_emb[:, 0], q_zh_emb[:, 1], c="red", alpha=alpha)
    
    plt.show()

q_en = questions["en"]["embeds"]
q_zh = questions["zh"]["embeds"]
c_en = candidates["embeds"][index["en"][0]:index["en"][1], :]
c_zh = candidates["embeds"][index["zh"][0]:index["zh"][1], :]

q_en_lir = lir(q_en, lan_info["en"])
c_en_lir = lir(c_en, lan_info["en"])
q_zh_lir = lir(q_zh, lan_info["zh"])
c_zh_lir = lir(c_zh, lan_info["zh"])
```

### The original system (X-X) exhibits the weak-aligment property

```python
pca_plot(q_en, c_en, q_zh, c_zh)
```

### After applying LIR, the multilingual system now is strongly aligned.

```python
pca_plot(q_en_lir, c_en_lir, q_zh_lir, c_zh_lir)
```
